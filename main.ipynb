```python
class HourlyDemandDataset:
    def __init__(self, path):
        self.path = path

    def load_data(self):
        df = pd.read_parquet(self.path)

        # Keep only necessary columns
        df = df[["tpep_pickup_datetime", "PULocationID"]]

        # Ensure datetime format
        df["tpep_pickup_datetime"] = pd.to_datetime(df["tpep_pickup_datetime"])

        # Extract hourly timestamp
        df["hour"] = df["tpep_pickup_datetime"].dt.floor("H")

        # Compute demand per hour per zone
        hourly = df.groupby(["hour", "PULocationID"]).size().reset_index(name="demand")

        # Sort for lag features
        hourly = hourly.sort_values(["PULocationID", "hour"]).reset_index(drop=True)

        return hourly
    
```

```python

def build_full_grid(hourly):
    all_hours = pd.date_range(start=hourly["hour"].min(),
                              end=hourly["hour"].max(),
                              freq='H')
    all_zones = hourly["PULocationID"].unique()

    full_index = pd.MultiIndex.from_product(
        [all_hours, all_zones],
        names=["hour", "PULocationID"]
    )

    full_df = full_index.to_frame().reset_index(drop=True)

    full_df = full_df.merge(hourly, on=["hour", "PULocationID"], how="left")
    full_df["demand"] = full_df["demand"].fillna(0)

    return full_df
```

```python
file_paths = [
    "data/yellow_tripdata_2023-01.parquet",
    "data/yellow_tripdata_2023-02.parquet",
    "data/yellow_tripdata_2023-03.parquet",
    "data/yellow_tripdata_2023-04.parquet",
    "data/yellow_tripdata_2023-05.parquet",
    "data/yellow_tripdata_2023-06.parquet",
    "data/yellow_tripdata_2023-07.parquet",
    "data/yellow_tripdata_2023-08.parquet",
    "data/yellow_tripdata_2023-09.parquet",
    "data/yellow_tripdata_2023-10.parquet",
    "data/yellow_tripdata_2023-11.parquet",
    "data/yellow_tripdata_2023-12.parquet"
]


all_monthly_hourly = []

for path in file_paths:
    dataset = HourlyDemandDataset(path)
    hourly_df = dataset.load_data()
    all_monthly_hourly.append(hourly_df)

# Combine all months
hourly_demand_all = pd.concat(all_monthly_hourly, ignore_index=True)

# ---- FIX: FILTER ONLY 2023 ----
hourly_demand_all = hourly_demand_all[
    hourly_demand_all['hour'].dt.year == 2023
]

# Re-group AFTER filtering
hourly_demand_all = (
    hourly_demand_all.groupby(["hour", "PULocationID"], as_index=False)
                     .agg({"demand": "sum"})
                     .sort_values(["PULocationID", "hour"])
                     .reset_index(drop=True)
)

print("Unique hours:", hourly_demand_all['hour'].nunique())
print("Row count before building grid:", hourly_demand_all.shape)

```

```python

full_grid_ = build_full_grid(hourly_demand_all)
print(full_grid_.shape)
```

```python
def add_lag_features(df):
    df = df.sort_values(["PULocationID", "hour"]).reset_index(drop=True)

    df["lag_1"] = df.groupby("PULocationID")["demand"].shift(1)
    df["lag_24"] = df.groupby("PULocationID")["demand"].shift(24)
    df["lag_168"] = df.groupby("PULocationID")["demand"].shift(168)

    return df
```

```python
import numpy as np
import pandas as pd
from pandas.tseries.holiday import USFederalHolidayCalendar

def build_features(df):
    """
    Input df must contain:
    - hour (datetime)
    - PULocationID
    - demand
    - lag_1, lag_24, lag_168
    - cluster_id
    """

    # ---------- BASIC TIME FEATURES ----------
    df["hour_of_day"] = df["hour"].dt.hour
    df["day_of_week"] = df["hour"].dt.dayofweek
    df["day_of_month"] = df["hour"].dt.day
    df["month"] = df["hour"].dt.month

    # ---------- FLAGS ----------
    df["is_weekend"] = df["day_of_week"].isin([5, 6]).astype(int)
    df["is_peak_hour"] = df["hour_of_day"].isin([7, 8, 9, 16, 17, 18]).astype(int)

    # ---------- CYCLICAL ENCODING ----------
    df["hour_sin"] = np.sin(2 * np.pi * df["hour_of_day"] / 24)
    df["hour_cos"] = np.cos(2 * np.pi * df["hour_of_day"] / 24)

    df["dow_sin"] = np.sin(2 * np.pi * df["day_of_week"] / 7)
    df["dow_cos"] = np.cos(2 * np.pi * df["day_of_week"] / 7)

    # ---------- HOLIDAY FEATURE ----------
    cal = USFederalHolidayCalendar()
    holidays = cal.holidays(
        start=df["hour"].min(),
        end=df["hour"].max()
    )

    df["is_holiday"] = df["hour"].dt.normalize().isin(holidays).astype(int)

    # ---------- LOG TRANSFORM (CRITICAL) ----------
    for col in ["demand", "lag_1", "lag_24", "lag_168"]:
        df[col] = np.log1p(df[col])

    # ---------- DROP RAW TIMESTAMP ----------
    df = df.drop(columns=["hour"])

    return df
```
